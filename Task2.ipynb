# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df
```

```python
#no.of rows & columns
iris_df.shape
```

```python
iris_df.info()
```

```python
x=iris_df.iloc[:,[0,1,2,3]].values
```

```python
#elbow method
from sklearn.cluster import KMeans

wcss =[]
for i in range(1,10):
    kmeans =KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
noc = range(1,10)
plt.plot(noc,wcss)
plt.title("The Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.show()
```


so k=3 is the optium no.of clusters from the above graph by The Elbow Method

# K-means Clustering

```python
k=3
k_means = KMeans(init="k-means++",n_clusters=k,n_init=12)
k_means.fit(x)
labels = k_means.labels_
iris_df["Clusters"]=labels
iris_df.head()
```

```python
plt.scatter(iris_df["sepal length (cm)"],iris_df["sepal width (cm)"],c=iris_df["Clusters"],cmap='rainbow')
```
